# Criando um Data Frame usando dicionário

```python
dados_dic = { 
    'Date' : ['2020-01-01' 
              , '2020-02-01' 
              , '2020-03-01' 
              , '2020-04-01' 
              , '2020-05-01'] ,
    'Temperatura' : [29.1 , 31.2 , 28.5 , 28.0 , 24.0] ,
    'Classification' : ["Quente" , "Muito quente" , "Quente" , "Quente" , "Confortável"]
}

display(dados_dic)
```

```python
dataframe = pd.DataFrame(dados_dic)

display(dataframe)
print(dataframe)
```

# Criando DataFrame apartir de um arquivo CSV

- O SCV significa "Valores Separados por Vírgula". Basicamente, são arquivos de texto onde cada linha representa a entrada de um valor, esses valores são separados por um caractere delimitador, geralmente é linha. 
Quando passamos esse arquivo para o pandas, ele consegue interpretar essas vírgulas como delimitadores e organizar as informações em linhas e colunas corretamente

```python
display(dataframe)

pd.DataFrame.to_csv(dataframe , "Arquivo.csv" , sep= "," , index= False)
pd.DataFrame.to_csv(dataframe , "Arquivo.csv2" , sep= "," , index= False)
```
